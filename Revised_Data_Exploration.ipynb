```python
#Lotfrontage is linear feet of street connected with property.So if we group LotFrontage by Neighborhood we can replace
#the missing values.
#What we have done here is used the fact that if the property is in the same neighborhood then taking the median of 
#LotFronatge of all the properties in tha neighborhood would serve as a good parameter for filling the missing LotFrontage
#in that neighborhood
lotfrontage_by_neighborhood=combined_data["LotFrontage"].groupby(combined_data["Neighborhood"])
for key,group in lotfrontage_by_neighborhood:
    idx=(combined_data["Neighborhood"]==key) & (combined_data["LotFrontage"].isnull())
    combined_data.loc[idx,"LotFrontage"]=group.median()
    
    
    
```

### Markers regarding creating new quality features

**1. Some features indicate quality of a particular facility.These are 
   ExterQual,ExterCond,BsmtQual,BsmtCond,HeatingQC,KitchenQual,FireplaceQu,GarageQual,GarageCond,PoolQC,BsmtExposure
   BsmtFinType1,BsmtFinType1,Functional,GarageFinish,FenceQuality**
   
**2. These features can be expressed in form of numbers depending upon the quality dscribed.See above feature analysis      for more clarification.(None->0,Ex->5,Gd->4,TA->3,Fa->2,Po->1).**

**3. We will make new quality features by replacing (None->0,Ex->5,Gd->4,TA->3,Fa->2,Po->1).On some trials I found that keeping both the features(new integral quality features and old string quality features) gave the best results.**

**4. Replace all the NA values suitably before applying the map function as applied below.The None in map function does not identifies the NA values in dataset.**

**5. Also be careful while using map function.Make sure all the values in your column are present in the mapping dictionary.Any unspecified value that map function encounters,it replaces it with NA.This gave me a lot of headache.**


 


### Markers regarding replacing NA values


**1. On observing the dataset one can easily conclude that all the garage related features have the same fraction of missing values.Also all the basement related features too have similar fraction of missing values.Nearly all these missing values(one out of set of Garage and basement related missing values)take place in the same observation.It can be safely assumed that these properties do not have garage or a basement appropriately.**

**2. Thus compensate for these missing values they can be replaced by string such as "NoGarage","NoBasement" etc.**

**3. We also need to create our quality features from some of these features.Instead of replacing missing values with "NoGarage" or "NoBasement" we can replace the values with a 0.This will not only facilitate creation of features but will create no difference whatsoever in dummy variable creation at the end as a feature of sorts "Feature_0" will be created.**

**4. All the other missing values are replaced by the most common value of that column.**

```python

#Replacing 'Utilities' missing values with most frequent value
combined_data['Utilities'].replace(np.nan,"AllPub",inplace=True)
    
#quality dictionary
qual_dict={'None':0,'Ex':5,'Gd':4,'TA':3,'Fa':2,'Po':1,0:0}

#mapping quality dictionary to create quality features
for feature in ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond','PoolQC']:
    #replacing missing values with 0
    combined_data[feature].fillna(0,inplace=True)
    #creating quality feature
    combined_data[feature+"Quality"]=combined_data[feature].map(qual_dict)

#Handling some more missing values before mapping
for feature in ['BsmtExposure','BsmtFinType1','BsmtFinType2','Functional','GarageFinish','Fence','Alley']:
    combined_data[feature].fillna(0,inplace=True)
    
    
#mapping suitable quality dictionary to different features to create quality features 
combined_data["BsmtExposureQuality"] = combined_data["BsmtExposure"].map({None: 0, "No": 1, "Mn": 2, "Av": 3, "Gd": 4,0:0})
combined_data["BsmtFinType1Quality"] = combined_data["BsmtFinType1"].map({None: 0, "Unf": 1, "LwQ": 2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6,0:0})
combined_data["BsmtFinType2Quality"] = combined_data["BsmtFinType2"].map({None: 0, "Unf": 1, "LwQ": 2, "Rec": 3, "BLQ": 4, "ALQ": 5, "GLQ": 6,0:0})
combined_data["FunctionalQuality"] = combined_data["Functional"].map({None: 0, "Sal": 1, "Sev": 2, "Maj2": 3, "Maj1": 4,"Mod": 5, "Min2": 6, "Min1": 7, "Typ": 8,0:0})
combined_data["GarageFinishQuality"] = combined_data["GarageFinish"].map({None: 0, "Unf": 1, "RFn": 2, "Fin": 3,0:0})
combined_data["FenceQuality"] = combined_data["Fence"].map({None: 0, "MnWw": 1, "GdWo": 2, "MnPrv": 3, "GdPrv": 4,0:0})
combined_data['AlleyQuality']=combined_data["Alley"].map({None:0,"Pave":1,"Grvl":2,0:0})
combined_data["StreetQuality"]=combined_data["Street"].map({None:0,"Pave":1,"Grvl":2,0:0})
combined_data["LotShapeQuality"]=combined_data["LotShape"].map({None:0,"Reg":3,"IR1":2,"IR2":1,"IR3":0})
combined_data["UtilitiesQuality"]=combined_data["Utilities"].map({None:0,"AllPub":3,"NoSewr":2,"NoSeWa":1,"ELO":0})




```

# Creating some more features

** Till now we have created quality features from those categorical features which were described qualitatively(good,fair,poor etc.).**

** We can now also make new quality features from other categorical features.**

** The approach used here is grouping "SalePrice" with respect to catagorical variables.Each catagory in each categorical variable would have a median sale price.Dictionary of all such medians for a categorical variables are formed.These medians are divided by the maximum of medians of this dictionary.If this operation yields value less than 0.2,the value is changed to 1,if in between 0.2 and 0.5 and so on.(Intuitively dividing as poor,average,good,great).**

```python
#FeatureEngineering
#Neighborhood

def process_dictionary(diction):
    maximum=max(diction.items(),key=operator.itemgetter(1))[1]
    for key,value in diction.items():
        diction[key]=diction[key]/maximum
        if diction[key]<=0.2:
            diction[key]=1
        elif diction[key]>0.2 and diction[key]<=0.5:
            diction[key]=2
        elif diction[key]>0.5 and diction[key]<=0.8:
            diction[key]=3
        elif diction[key]>0.8:
            diction[key]=4
    return diction





median_table=train_data["SalePrice"].groupby(train_data["Neighborhood"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["NeighborhoodQuality"]=combined_data["Neighborhood"].map(median_dict)
#MSZoning
median_table=train_data["SalePrice"].groupby(train_data["MSZoning"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["MSZoningQuality"]=combined_data["MSZoning"].map(median_dict)
#ContourQuality
median_table=train_data["SalePrice"].groupby(train_data["LandContour"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["LandContourQuality"]=combined_data["LandContour"].map(median_dict)

median_table=train_data["SalePrice"].groupby(train_data["LotConfig"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["LotConfigQuality"]=combined_data["LotConfig"].map(median_dict)

median_table=train_data["SalePrice"].groupby(train_data["LandSlope"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["LandSlopeQuality"]=combined_data["LandSlope"].map(median_dict)

median_table=train_data["SalePrice"].groupby(train_data["BldgType"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["BldgTypeQuality"]=combined_data["BldgType"].map(median_dict)

median_table=train_data["SalePrice"].groupby(train_data["HouseStyle"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["HouseStyleQuality"]=combined_data["HouseStyle"].map(median_dict)

median_table=train_data["SalePrice"].groupby(train_data["RoofStyle"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["RoofStyleQuality"]=combined_data["RoofStyle"].map(median_dict)


median_table=train_data["SalePrice"].groupby(train_data["RoofMatl"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["RoofMatlQuality"]=combined_data["RoofMatl"].map(median_dict)


median_table=train_data["SalePrice"].groupby(train_data["MasVnrType"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["MasVnrTypeQuality"]=combined_data["MasVnrType"].map(median_dict)


median_table=train_data["SalePrice"].groupby(train_data["Foundation"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["FoundationQuality"]=combined_data["Foundation"].map(median_dict)


median_table=train_data["SalePrice"].groupby(train_data["Electrical"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["ElectricalQuality"]=combined_data["Electrical"].map(median_dict)


median_table=train_data["SalePrice"].groupby(train_data["SaleCondition"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["SaleConditionQuality"]=combined_data["SaleCondition"].map(median_dict)

median_table=train_data["SalePrice"].groupby(train_data["MoSold"]).median()
median_dict=dict(median_table)
median_dict=process_dictionary(median_dict)
combined_data["MoSoldQuality"]=combined_data["MoSold"].map(median_dict)



```
